# feature selection

```python
movies.columns    #displays columns
```

```python
movies=movies[['id', 'title', 'genre', 'overview']]
```

```python
movies
```

```python
movies['tags']= movies['overview'] + movies['genre']    #created a single tag as content based RS
```

```python
movies
```

```python
new_data=movies.drop(columns=['overview', 'genre'])   #dropping the columns after making them into a single tag
```

```python
new_data
```

# you have to convert the tags into numerical representation so use scikit learn

```python
from sklearn.feature_extraction.text import CountVectorizer   #countvectorizer is used to convert into vectors
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)    #calculates the distance between two if distance is less they are more similar
```

```python
similarity
```

```python
new_data[new_data['title']=='The Godfather'].index[0]
```

```python
new_data.info()
```

```python
distance =sorted(list(enumerate(similarity[2])), reverse=True, key= lambda vector: vector[1])   # reverse sorted gives the lowest similarity distance which means they are more similar
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance =sorted(list(enumerate(similarity[index])), reverse=True, key= lambda vector: vector[1])   # reverse sorted gives the lowest similarity distance which means they are more similar
    for i in distance[0:5]:
      print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
